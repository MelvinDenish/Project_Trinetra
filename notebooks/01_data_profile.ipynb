# 01 — Data Profile (Phase 0)

Reproduces the headline findings that drive the architecture: the **skills/education decoy**, the geography skew, the unreliable `years_of_experience`, the trap populations, and the honeypot counts.

Point `CANDIDATES` at your `candidates.jsonl`. The same logic ships as `scripts/profile_data.py`.

```python
import sys, collections, statistics
sys.path.insert(0, '../src')
from redrob_ranker import io_utils
CANDIDATES = '../data/sample_candidates.jsonl'  # replace with the full candidates.jsonl
cands = io_utils.load_candidates(CANDIDATES)
len(cands)
```

## Skills are a decoy — near-uniform frequency, decoupled from role

```python
skills = collections.Counter()
for c in cands:
    for s in c.get('skills', []):
        skills[s['name']] += 1
print('unique skill names:', len(skills))
skills.most_common(10)
```

## `years_of_experience` is unreliable — derive from career history instead

```python
stated = [io_utils.stated_yoe(c) for c in cands]
derived = [io_utils.derived_experience_years(c) for c in cands]
print('stated median :', round(statistics.median(stated), 2))
print('derived median:', round(statistics.median(derived), 2))
```
