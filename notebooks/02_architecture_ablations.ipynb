## Inspect the ranking interactively
Load artifacts + candidates and explore the sub-scores behind any candidate.

```python
import os, sys
os.environ['HF_HUB_OFFLINE'] = '1'; os.environ['TRANSFORMERS_OFFLINE'] = '1'
sys.path.insert(0, '../src')
from redrob_ranker import pipeline
emb, ids, bm25 = pipeline.load_artifacts()
cands = pipeline._align_candidates('../candidates.jsonl', ids)
final, detail = pipeline._score_all(cands, emb, bm25, log=print)
import numpy as np
order = np.argsort(-final)[:10]
for i in order:
    p = cands[i]['profile']
    print(round(float(final[i]),3), cands[i]['candidate_id'], '|', p.get('current_title'), '|', p.get('country'))
```
